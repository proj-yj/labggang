```python
import glob
from ase.io import read,write
from ase import Atoms,Atom
from collections import Counter
import numpy as np
from tqdm import tqdm
from make_representation import *
import pickle
```

```python
cif_path = './cif'
#glob 모듈의 glob 함수는 사용자가 제시한 조건에 맞는 파일명을 리스트 형식으로 반환한다.
cif_list = glob.glob(cif_path+'/*.cif')

results = []
name_list = [] 

for i in cif_list:
    name = i.split(',')[0]
    atoms =read(i)
    s = atoms.get_chemical_symbols()
    n_fe = s.count('Fe')
    n_o = s.count('O')
    print(n_fe, n_o)
    print(atoms)
    if n_fe > 40 or n_o > 40:
        continue
    else:
        image = do_feature(atoms)
        print(image)

        results.append(image)
        name_list.append(name)


results = np.array(results)
# print(results)

print(results.shape)
np.save('unique_FeO',results)
#name_list는 파일명이다.
with open("unique_FeO_name_list",'wb') as f:
    pickle.dump(name_list,f)
```
