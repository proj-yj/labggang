```python
import os
import glob
import numpy as np
from ase.io import read, write
from ase import Atoms,Atom
import view_atoms_gan
import itertools
import make_representation
from tqdm import tqdm
import pickle
```

```python
def do_supercell(image, n = None):
    atoms, image = view_atoms_gan.view_atoms(image, view = False)
    # s  = atoms.get_chemical_symbols()
    # n_mg = s.count("Mg")
    # n_mn = s.count('Mn')
    # n_o = s.count('O')

    atoms_x = atoms.repeat((2,1,1))
    atoms_y = atoms.repeat((1,2,1))
    atoms_z = atoms.repeat((1,1,2))
    
    image_x = make_representation.do_feature(atoms_x).reshape(1,82,3)
    image_y = make_representation.do_feature(atoms_y).reshape(1,82,3)
    image_z = make_representation.do_feature(atoms_z).reshape(1,82,3)
    
    new_images = np.vstack((image_x,image_y,image_z))
    return new_images
```
